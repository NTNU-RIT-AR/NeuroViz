```python
from neuroviz import NeuroViz, ParameterDict, default_parameters
```

```python
neuro = NeuroViz(port=9001, use_secret=False)
```

```python
default_parameters()
```

```python
neuro.set_live_parameters({
  'transparency': 0.5,
  'glow': 0.0,
  'smoothness': 0.5,
  'emission': 0.0,
  'light_intensity': 1.0,
  'light_temperature': 6500.0
})
```

```python
a: ParameterDict = {
  **default_parameters(),
  'transparency': 1.0,
}

b = {
  **default_parameters(),
  'transparency': 0.5,
}

chosen = neuro.prompt_choice(a, b)
print(chosen)
```

```python
a: ParameterDict = {
  **default_parameters(),
  'glow': 1.0,
}

rating = chosen = neuro.prompt_rating(a)
print(rating)
```
